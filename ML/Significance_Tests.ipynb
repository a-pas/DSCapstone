```python
import pandas as pd
import statsmodels.api as sm
from sklearn.preprocessing import OneHotEncoder

# Load the dataset
df1 = pd.read_csv("/content/combined_data_2024.csv")

# Encode the categorical variable
df1['OFFENSE'] = df1['OFFENSE'].astype('category')
df1['OFFENSE'] = df1['OFFENSE'].cat.codes  # Convert categorical to numeric

# Define independent variables (numerical features)
X = df1[['YEAR', 'MONTH', 'DAY', 'HOUR', 'MINUTE', 'WARD', 'DISTRICT', 'PSA', 'LATITUDE', 'LONGITUDE']]
X = sm.add_constant(X)  # Add constant for the intercept
y = df1['OFFENSE']

# Fit the multinomial logistic regression model
model = sm.MNLogit(y, X)  # Use MNLogit for multinomial logistic regression
result = model.fit()

# Print the summary to see p-values
print(result.summary())
```

```python
from scipy.stats import kruskal

# List to store significant results
significant_features = []

# Perform Kruskal-Wallis H test for each numerical column
for col in ['YEAR', 'MONTH', 'DAY', 'HOUR', 'MINUTE', 'WARD', 'DISTRICT', 'PSA', 'LATITUDE', 'LONGITUDE']:
    groups = [group[col].values for name, group in df1.groupby('OFFENSE')]
    stat, p_value = kruskal(*groups)

    if p_value < 0.05:  # Check if the p-value is less than 0.05
        significant_features.append((col, stat, p_value))

# Print significant features
print("Significant numerical features (p < 0.05):")
for feature in significant_features:
    print(f"Feature: {feature[0]}, Statistic: {feature[1]}, p-value: {feature[2]}")
```
